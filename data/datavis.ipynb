```python
import pandas as pd
import numpy as np
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
root = Path("./")
```

```python
df = pd.read_csv(root / "20231204v1.csv")
```

```python
fig, ax = plt.subplots(figsize=(6, 4), dpi=150)
sns.histplot(ax=ax, data=df, x="Ionic conductivity (S cm-1)", 
             bins=100, kde=True)
# ax.set_xscale('log')
```

```python
print("Mean: {:.8f}".format(df["Ionic conductivity (S cm-1)"].mean()))
print("Median: {:.8f}".format(df["Ionic conductivity (S cm-1)"].median()))
print("Std: {:.8f}".format(df["Ionic conductivity (S cm-1)"].std()))
print("Minimum: {:.8f}".format(df["Ionic conductivity (S cm-1)"].min()))
print("Maximum: {:.8f}".format(df["Ionic conductivity (S cm-1)"].max()))
```
